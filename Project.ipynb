```python
days = 365

dt = 1/days

mu = rets.mean()['WIT']

sigma = rets.std()['WIT']
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    price= np.zeros(days)
    price[0]=start_price
    
    shock=np.zeros(days)
    drift=np.zeros(days)
    
    
    for x in range(1, days):
        
        shock[x]=np.random.normal(loc= mu*dt, scale=sigma* np.sqrt(dt))
        drift[x]=mu* dt
        
        price[x]=price[x-1]+ (price[x-1]* (shock[x]+drift[x]))
        
        
    return price    
    
```

```python
INFY.describe()
```

```python
start_price= 14.710000
for run in range( 100):
    plt.plot(stock_monte_carlo(start_price,days,mu,sigma))
plt.xlabel("Days")
plt.ylabel("Price") 
plt.legend(['green','pink'])
```

```python
runs=50
simulations = np.zeros(runs)
np.set_printoptions(threshold=5)

for run in range(run):
    simulations[run]= stock_monte_carlo(start_price,days,mu,sigma)[days-1]
```

```python
# Now we'lll define q as the 1% empirical qunatile, this basically means that 99% of the values should fall between here
q = np.percentile(simulations, 1)
plt.hist(simulations,bins=100)

plt.figtext(0.2, 0.5, s="start price= $ %.2f" %start_price)
plt.figtext(0.2, 0.4, s="Mean price= $ %.2f" %simulations.mean())
plt.figtext(0.2, 0.3, s="VaR(0.99):  $ %.2f" %(start_price- q))
plt.figtext(0.2, 0.2, s="q(0.99)  $ %.2f" % q)


#plt.axvline(x=q, linewidth=4, color='r')

plt.title(u"Final price distribution for INFY Stock after %s days" % days, weight='bold')

```
